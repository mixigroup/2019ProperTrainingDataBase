## 4. GROUP BY å¥

`GROUP BY` å¥ã‚’ä½¿ã†ã¨æŒ‡å®šã—ãŸã‚«ãƒ©ãƒ ã®å€¤ã§ã‚°ãƒ«ãƒ¼ãƒ—åˆ†ã‘ã—ãŸçµæœã«å¯¾ã—ã¦é›†è¨ˆã‚’è¡Œã†ã“ã¨ãŒã§ãã¾ã™ï¼ˆè¤‡æ•°ã®ã‚«ãƒ©ãƒ ã‚’æŒ‡å®šã§ãã¾ã™ï¼‰ã€‚ä»¥ä¸‹ã¯ `titles` ãƒ†ãƒ¼ãƒ–ãƒ«ã‹ã‚‰ `title` ã”ã¨ã«ãƒ¬ã‚³ãƒ¼ãƒ‰æ•°ã‚’æ•°ãˆã¦å¤šã„é †ã«ä¸¦ã¹æ›¿ãˆã‚‹ã‚¯ã‚¨ãƒªã§ã™ã€‚

````python
%%sql
select
    title,
    count(1) as cnt
from
    titles
group by
    title

order by
    cnt desc
````

## 5. HAVING å¥

HAVING å¥ã«æ¡ä»¶ã‚’è¨˜è¿°ã™ã‚‹ã“ã¨ã§é›†è¨ˆå¾Œã®çµæœã«å¯¾ã—ã¦æ¤œç´¢ã‚’ã™ã‚‹ã“ã¨ãŒã§ãã¾ã™ã€‚

````python
%%sql
select
    title,
    count(1) as cnt
from
    titles
group by
    title

-- "where cnt < 20000" ã¯ã‚¨ãƒ©ãƒ¼
having
    cnt < 20000
    
order by
    cnt desc
````
